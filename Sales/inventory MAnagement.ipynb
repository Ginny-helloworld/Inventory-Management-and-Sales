```python
import time
fd=open('Inventory.txt','r')
products=fd.read().split('\n')
fd.close()

ui_name=    input("Enter your Name.            :")
ui_Email_ID=input("Enter your email-ID.        :")
ui_ph=      input("Enter your phone number     :")
ui_id=      input("Enter the product Id.       :")
ui_qty=     input("Enter the product qty.      :")

updated_prod=[]

for product in products:
    prod_details=product.split(',')
    
    if prod_details[0]==ui_id:
        
        if int(prod_details[3])>=int(ui_qty):
            
            print("-----------------------------------------------")
            print("                      Bill                     ")        
            print("Porduct ID.              :", prod_details[0])
            print("Product Name             :", prod_details[1])
            print("Product Price            :", prod_details[2])
            print("Quantity                 :", ui_qty)
            print("-----------------------------------------------")
            print("Total                    :", int(ui_qty)*int(prod_details[2]))
            print("-----------------------------------------------")
            
            prod_details[3]=str(int(prod_details[3])-int(ui_qty))
            
            fd=open('Sales.txt','a')
            sales_details=ui_name+"||"+ui_Email_ID+"||"+ui_ph+"||"+ui_id+"||"+ui_qty+"||"+prod_details[2]+"||"+str(int(ui_qty)*int(prod_details[2]))+"||"+time.ctime()+"||"+"\n"            
            fd.write(sales_details)
            fd.close()
            
            
            
        elif prod_details=='0':
            print("Sorry! item out of stock")
        
        
        else:
            print("Sorry! We are not having enough quantity")
            print("We are having only", prod_details[3],"Quantity")
            print("Would you like to purchase it?")
            
            ch=input("Press Y/N")
            
            if ch=='Y' or ch =='y':
            
                print("-----------------------------------------------")
                print("                      Bill                     ")        
                print("Porduct ID.              :", prod_details[0])
                print("Product Name             :", prod_details[1])
                print("Product Price            :", prod_details[2])
                print("Quantity                 :", prod_details[3])
                print("-----------------------------------------------")
                print("Total                    :", int(prod_details[3])*int(prod_details[2]))
                print("-----------------------------------------------")
                
                fd=open('Sales.txt','a')
                sales_details=ui_name+"|"+ui_Email_ID+"|"+ui_ph+"|"+ui_id+"|"+prod_details[2]+"|"+str(int(prod_details[3])*int(prod_details[2]))+"||"+time.ctime()+"||"+"\n"                                                                                                     
                fd.write(sales_details)
                fd.close()
                
                prod_details='0'
                
            else:
                print("Thankyou for visiting")
            
        
            
    updated_prod.append(prod_details)
    
lst = []

for i in updated_prod:
    if len(i) >= 4:
        prod = i[0] + "," + i[1] + "," + i[2] + "," + i[3] + "\n"
        lst.append(prod)

# Remove the newline character from the last line
if lst:
    lst[-1] = lst[-1][:-1]


fd=open('Inventory.txt','w')

for i in lst:  
    fd.write(i)   
fd.close()

                
        
   
print("Thankyou")   
```
